Day 1, Puzzle 1

```python
# Define the file path
data_file = 'data/input.txt' 

# Open the file in read mode and read its lines
with open(data_file, 'r') as input_file: 
    lines = input_file.read().splitlines() # Split the file in to a list of lines
```

```python
# Read file to see data
lines 
```

```python
# Convert each line in the list to an integer
lines = [int(line) for line in lines]
```

```python
# Read lines again to see if the data changed properly or not
lines
```

# Puzzle 1 solution

```python
# Iterate through each pair of numbers in 'lines'
for a in lines:
    for b in lines:
        if a + b == 2020:            # Check if the sum of the pair is 2020
            print(a,b, a + b,a * b)  # Print the pair and their sum/product
            solution = a * b         # Store the product of the pair
print(f'Puzzle 1: {solution}')       # Print the solution for Puzzle 1
```

# Puzzel 2 solution

```python
# Iterate through each pair of numbers in 'lines'
for a in lines:
    for b in lines:
        for c in lines:
            if a + b +c == 2020:                       # Check if the sum of the pair is 2020
                print(a,b,c, a + b + c,a * b * c)      # Print the pair and their sum/product
                solution = a * b * c                   # Store the product of the pair
print(f'Puzzle 2: {solution}')                         # Print the solution for Puzzle 2
```

# Puzzel 1 solution using pairs

```python
pairs = [(a, b) for a in lines for b in lines if a + b == 2020]    # Find all pairs of numbers that sum to 2020
pair =pairs[0]                                                     # Select the first pair from the list of pairs
pair                                                               # Display the selected pair
```

# Puzzel 2 solution using pairs

```python
pairs = [(a, b, c) for a in lines for b in lines for c in lines if a + b + c == 2020]   # Find all pairs of numbers that sum to 2020
pair =pairs[0]                                                                          # Select the first pair from the list of pairs
pair                                                                                    # Display the selected pair
```
