# Testing

```python
# Define the file path
data_file_test = 'data/input_test.txt' 

# Open the file in read mode and read its lines
with open(data_file_test, 'r') as input_file_test: 
    lines = input_file_test.read().splitlines() # Split the file in to a list of lines
```

```python
# Convert each line in the list to an integer
lines = [int(line) for line in lines]
```

```python
# Read file to see data
lines 
```

```python
# set input variable depths
depths = lines

# Initialize a counter for the number of increases
window_increases = 0

# Loop through the depths and calculate the sum of each three measurement window
for i in range(2, len(depths)):  # Start at index 2 
    
    # Sum of the current and the previous two measurements
    current_window_sum = depths[i] + depths[i - 1] + depths[i - 2]
    
    # Compare it with the previous window sum 
    if i > 2:  # skip the first window (A) since there's no previous sum
        previous_window_sum = depths[i - 1] + depths[i - 2] + depths[i - 3] # cvheck if the current windo sum is greater than the previous window sum
        if current_window_sum > previous_window_sum: # if the current window sum is greatert, increase the coutner by 1
            window_increases += 1

# Output the result
print(f"Number of increases in the three-measurement window sums: {window_increases}")
```
